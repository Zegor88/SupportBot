```python
import json
import pandas as pd
from datetime import datetime

# Читаем историю диалогов
with open('../data/dialog_history.json', 'r', encoding='utf-8') as f:
    history = json.load(f)

# Получаем данные пользователя
user_id = '284544528'  # ID пользователя для анализа
user_data = history[user_id]

print(f"\nПользователь: @{user_data['user_info']['username']} ({user_data['user_info']['user_name']})")
print('=' * 80)

# Выводим каждый диалог
for dialog in user_data['dialogs']:
    print(f"\nДата: {dialog['date']}")
    print(f"Сообщение: {dialog['message']}")
    
    # Выводим контекст, если есть
    if dialog.get('context'):
        print("\nКонтекст RAG:")
        print(dialog['context'])
    
    # Выводим ответ бота, если есть
    if dialog.get('bot_response'):
        print(f"\nОтвет бота ({dialog['bot_response']['date']}):")
        print(dialog['bot_response']['text'])
    
    print('\n' + '-' * 80)

# Создаем DataFrame для анализа
dialogs_data = []

for dialog in user_data['dialogs']:
    dialog_info = {
        'Дата': dialog['date'],
        'Сообщение': dialog['message'],
        'Есть контекст': 'Да' if dialog.get('context') else 'Нет',
        'Есть ответ': 'Да' if dialog.get('bot_response') else 'Нет',
    }
    if dialog.get('bot_response'):
        dialog_info['Время ответа'] = dialog['bot_response']['date']
    dialogs_data.append(dialog_info)

# Создаем и выводим DataFrame
df = pd.DataFrame(dialogs_data)
print("\nСтатистика диалогов:")
print("=" * 80)
print(df.to_string(index=False))

# Выводим общую статистику
print("\nОбщая статистика:")
print("=" * 80)
print(f"Всего сообщений: {len(df)}")
print(f"Сообщений с контекстом: {len(df[df['Есть контекст'] == 'Да'])}")
print(f"Сообщений с ответами: {len(df[df['Есть ответ'] == 'Да'])}") 
```
